```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['species'].unique()
```

```python
#1st of all seeing with train test split
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species',axis=1)
y = df['species']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=10)
```

```python
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma='auto')
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
#Now using random search CV
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#In Random search cv train test split is not needed, it happpens auto in the programme
```

```python
classifier = RandomizedSearchCV((model_svm),{
  'C':[1,10,20,30],
  'kernel':['rbf','linear'],
},n_iter=4,cv= 5,return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
#After using random search cv the model score
model_svm.score(X,y)
```
